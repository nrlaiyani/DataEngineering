```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import numpy as np

# List of URLs to scrape
urls = [
    "https://www.sephora.my/categories/skincare/cleanser-and-exfoliator/facial-cleanser",
    "https://www.sephora.my/categories/skincare/cleanser-and-exfoliator/facial-cleanser?page=2",
    "https://www.sephora.my/categories/skincare/cleanser-and-exfoliator/facial-cleanser?page=3",
]

# Headers for request
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}

# Initialize lists to store extracted data
all_brands = []
all_products = []
all_prices = []
all_sizes = []
all_reviews = []
all_product_images = []
all_ratings = []
all_where_to_buy = []

# Loop through all URLs
for url in urls:
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        page_soup = BeautifulSoup(response.content, 'html.parser')

        # Extract brands
        brands = page_soup.find_all("div", {"class": "brand"})
        all_brands.extend([brand.text.strip() for brand in brands])

        # Extract product names
        product_names = page_soup.find_all("div", {"class": "product-name"})
        all_products.extend([product.text.strip() for product in product_names])

        # Extract prices
        prices = page_soup.find_all("div", {"class": "prices"})
        all_prices.extend([price.text.strip() for price in prices])

        # Extract sizes
        sizes = page_soup.find_all("div", {"class": "variants-count"})
        all_sizes.extend([size.text.strip() for size in sizes])

        # Extract reviews
        reviews = page_soup.find_all("span", {"class": "reviews-count"})
        all_reviews.extend([review.text.strip() for review in reviews])

        # Extract product images
        product_images = page_soup.find_all("a", {"class": "product-card-image-link"})
        all_product_images.extend([product_image.get('href') for product_image in product_images])

        # Extract ratings (percentage) and convert to a star rating (1-5 scale)
        star_elements = page_soup.find_all("div", {"class": "stars"})
        ratings = []

        for star in star_elements:
            style = star.get("style", "")
            percentage = np.nan

            # Extract the percentage value from inline style
            if "--highlighted-percentage" in style:
                start_idx = style.find("--highlighted-percentage:") + len("--highlighted-percentage:")
                end_idx = style.find("%", start_idx)
                percentage_str = style[start_idx:end_idx].strip()

                if percentage_str:
                    percentage = float(percentage_str)

            # Convert percentage to star rating (1 to 5 scale)
            if 0 <= percentage <= 100:
                star_rating = (percentage / 100) * 4 + 1
                ratings.append(round(star_rating, 1))
            else:
                ratings.append(np.nan)

        # Append ratings to the main list
        all_ratings.extend(ratings)

        # Extract "Where to Buy" information
        where_to_buy = page_soup.find_all("span", {"class": "label"})
        all_where_to_buy.extend([buy.text.strip() for buy in where_to_buy])

        time.sleep(1)  # Pause to avoid getting blocked

    else:
        print(f"Failed to fetch {url}, Status Code: {response.status_code}")

# Ensure all lists have the same length
max_length = max(len(all_brands), len(all_products), len(all_prices), len(all_sizes),
                 len(all_reviews), len(all_product_images), len(all_ratings), len(all_where_to_buy))

all_brands += ["N/A"] * (max_length - len(all_brands))
all_products += ["N/A"] * (max_length - len(all_products))
all_prices += ["N/A"] * (max_length - len(all_prices))
all_sizes += ["N/A"] * (max_length - len(all_sizes))
all_reviews += ["N/A"] * (max_length - len(all_reviews))
all_product_images += ["N/A"] * (max_length - len(all_product_images))
all_ratings += ["N/A"] * (max_length - len(all_ratings))
all_where_to_buy += ["N/A"] * (max_length - len(all_where_to_buy))

# Create a DataFrame
data = {
    "Brand": all_brands,
    "Product Name": all_products,
    "Price": all_prices,
    "Size": all_sizes,
    "Review Count": all_reviews,
    "Product Image Link": all_product_images,
    "Rating (Out of 5)": all_ratings,
    "Where to Buy": all_where_to_buy,
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("sephora_products.csv", index=False)
print("Data saved to 'sephora_products.csv'")
```

Cleaning Task 2: Converting String types to numerical categories

```python
import pandas as pd

# Load the dataset, ensuring N/A-like values are treated as strings, not NaN
df = pd.read_csv("sephora_products.csv", keep_default_na=False)

# **1️⃣ Clean the "Size" column:** Extract numeric value and convert to float
df["Size"] = df["Size"].replace("N/A", "1 size")
df["Size"] = df["Size"].str.extract(r'(\d+)').astype(float)

# **2️⃣ Clean "Review Count" column:** Remove parentheses and convert to integers
df["Review Count"] = df["Review Count"].str.replace(r'[()]', '', regex=True).replace("N/A", "0")
df["Review Count"] = pd.to_numeric(df["Review Count"], errors='coerce').fillna(0).astype(int)

# **3️⃣ Clean the "Where to Buy" column:** Replace "N/A" with "To Be Confirmed"
df["Where to Buy"] = df["Where to Buy"].replace("N/A", "To Be Confirmed")

# **4️⃣ Clean the "Price" column:** Remove currency symbols and convert to numerical data
df["Price"] = df["Price"].replace({r"[^0-9.]": ""}, regex=True)
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# **5️⃣ Clean the "Rating (Out of 5)" column:** Convert to numeric, handle "N/A"
df["Rating (Out of 5)"] = df["Rating (Out of 5)"].replace("N/A", "No Rating")
df["Rating (Out of 5)"] = pd.to_numeric(df["Rating (Out of 5)"], errors="coerce")

# Save the cleaned dataset to a new CSV file
df.to_csv("sephora_products_cleaned.csv", index=False)

print("✅ All columns have been cleaned, including 'Rating'. Data saved to 'sephora_products_cleaned.csv'.")
```

VISUALISATION (After Cleaning)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the raw dataset without any cleaning
df_raw = pd.read_csv("sephora_products.csv", keep_default_na=False)

# **Fix Column Names**
df_raw.columns = df_raw.columns.str.strip().str.lower()  # Standardize column names

# **Replace "N/A" in 'rating (out of 5)' with 'No Rating' BEFORE replacing other N/A values**
df_raw.loc[df_raw["rating (out of 5)"] == "N/A", "rating (out of 5)"] = "No Rating"

# **Now, replace "N/A" with "To Be Confirmed" for all OTHER columns**
df_raw.replace("N/A", "To Be Confirmed", inplace=True)

# Convert numeric ratings while preserving "No Rating"
df_raw["rating_numeric"] = pd.to_numeric(df_raw["rating (out of 5)"], errors="coerce")  # Keep numeric ratings

# Set up the plot style
sns.set(style="whitegrid")

# ✅ **Plot 5 Histograms AFTER Cleaning**
fig, axes = plt.subplots(1, 5, figsize=(25, 5))

# 1️⃣ Histogram for 'Size' after cleaning
if 'size' in df_raw.columns:
    size_counts = df_raw['size'].value_counts()
    axes[0].bar(size_counts.index[:10], size_counts.values[:10], color='lightcoral')  # Top 10 sizes for readability
    axes[0].set_title("Histogram for Size (After Cleaning)")
    axes[0].set_xlabel('Size')
    axes[0].set_ylabel('Frequency')
    axes[0].tick_params(axis='x', rotation=45)
else:
    print("Warning: 'size' column not found in dataset!")

# 2️⃣ Histogram for 'Where to Buy' after cleaning
if 'where to buy' in df_raw.columns:
    where_counts = df_raw['where to buy'].value_counts()
    axes[1].bar(where_counts.index, where_counts.values, color='goldenrod')
    axes[1].set_title("Histogram for Where to Buy (After Cleaning)")
    axes[1].set_xlabel('Where to Buy')
    axes[1].set_ylabel('Frequency')
    axes[1].tick_params(axis='x', rotation=45)
else:
    print("Warning: 'where to buy' column not found in dataset!")

# 3️⃣ Histogram for 'Price' after cleaning
if 'price' in df_raw.columns:
    axes[2].hist(df_raw['price'], bins=20, color='lightblue', edgecolor='black')
    axes[2].set_title("Histogram for Price (After Cleaning)")
    axes[2].set_xlabel('Price')
    axes[2].set_ylabel('Frequency')
else:
    print("Warning: 'price' column not found in dataset!")

# 4️⃣ Histogram for 'Rating' after cleaning (including "No Rating")
if 'rating (out of 5)' in df_raw.columns:
    # Separate numeric ratings & "No Rating" count
    rating_counts = df_raw["rating (out of 5)"].value_counts()

    # Bar plot for rating distribution
    axes[3].bar(rating_counts.index, rating_counts.values, color='lightgreen')
    axes[3].set_title("Histogram for Rating (After Cleaning)")
    axes[3].set_xlabel('Rating (Out of 5)')
    axes[3].set_ylabel('Frequency')
    axes[3].tick_params(axis='x', rotation=45)
else:
    print("Warning: 'rating (out of 5)' column not found in dataset!")

# 5️⃣ Histogram for 'Review Count' after cleaning
if 'review count' in df_raw.columns:
    axes[4].hist(df_raw['review count'], bins=20, color='skyblue', edgecolor='black')
    axes[4].set_title("Histogram for Review Count (After Cleaning)")
    axes[4].set_xlabel('Review Count')
    axes[4].set_ylabel('Frequency')
else:
    print("Warning: 'review count' column not found in dataset!")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()

# Save the updated dataset
df_raw.to_csv("sephora_products_updated.csv", index=False)
print("✅ 'N/A' values replaced with 'To Be Confirmed' and 'No Rating'. Data saved as 'sephora_products_updated.csv'.")
```
